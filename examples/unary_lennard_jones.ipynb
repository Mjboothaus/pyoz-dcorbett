```python
import pyoz as oz
import matplotlib.pyplot as plt
plt.style.use('seaborn-colorblind')
%matplotlib inline
```

```python
lj_unary = oz.System()
lj_unary.set_interaction(0, 0, oz.lennard_jones(lj_unary.r, eps=1.0, sig=2.0))
r = lj_unary.r
U_r = lj_unary.U_r
```

```python
plt.plot(r, U_r[0, 0])
plt.xlabel('r')
plt.ylabel('U(r)')
plt.xlim(0, 10)
plt.ylim(-1.5, 1)
```

```python
g_r, c_r, e_r, h_k = lj_unary.solve(rhos=0.1, closure_name='hnc')
```

```python
plt.plot(r, g_r[0, 0])
plt.xlabel('r')
plt.ylabel('g(r)')
plt.xlim(0, 10)
```
